```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
df=pd.read_csv("/content/Iris 8.csv")
print(df.head(10))
```

```python
features=df.iloc[:,1:5]
print(features.head(5))
```

```python
labels=df.iloc[:,5]
print(labels.head(5))
```

```python
from sklearn.model_selection import train_test_split
features_train, features_test, labels_train, labels_test=train_test_split(features, labels, test_size=0.2, random_state=0)
print(features_train.info)
print(features_test.info)
print(labels_train)
print(labels_test)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
sc=StandardScaler()
sc.fit(features_train)
features_train_std=sc.transform(features_train)
features_test_std=sc.transform(features_test)
print(features_train_std)
print(features_test_std)
```

Support Vector Classifier

```python
from sklearn.svm import SVC
svm=SVC(kernel='rbf', random_state=0, gamma=.10, C=1.0)
svm.fit(features_train_std, labels_train)
print("The accuracy of SVC on training data is {:.2f}".format(svm.score(features_train_std,labels_train)))
print("The accuracy of SVC on testing data is {:.2f}".format(svm.score(features_test_std, labels_test)))
```

KNN Classifier(K nearest Neighbors)

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7, p=2, metric="minkowski")
knn.fit(features_train_std, labels_train)
print("The accuracy of KNN Classifier on training data is {:.2f}".format(svm.score(features_train_std, labels_train)))
print("The accuracy of KNN Classifier on testing data is {:.2f}".format(svm.score(features_test_std, labels_test)))
```

XGBoost Classifier

```python
import xgboost as xgb
xgb_cl=xgb.XGBClassifier()
xgb_cl=xgb_cl.fit(features_train_std, labels_train)
print("The accuracy of XGBoost Classifier on training data is {:.2f}".format(xgb_cl.score(features_train_std, labels_train)))
print("The accuracy of XGBoost Classifier on testing data is {:.2f}".format(xgb_cl.score(features_test_std, labels_test)))
```

Decision Tree

```python
from sklearn import tree
decision_tree=tree.DecisionTreeClassifier(criterion="gini")
decision_tree.fit(features_train_std,labels_train)
print("The accuracy of Decision Tree Classifier on training data is {:.2f}".format(decision_tree.score(features_train_std, labels_train)))
print("The accuracy of Decision Tree Classifier on testin data is {:.2f}".format(decision_tree.score(features_test_std, labels_test)))
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
random_forest=RandomForestClassifier()
random_forest.fit(features_train_std,labels_train)
print("The accuracy of Random Forest Classifier on training data is {:.2f}".format(random_forest.score(features_train_std, labels_train)))
print("The accuracy of Random Forest Classifier on testing data is {:.2f}".format(random_forest.score(features_test_std, labels_test)))
```
