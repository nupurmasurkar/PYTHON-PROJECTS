```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
%matplotlib inline
```

```python
data=pd.read_csv("/content/boston.csv")
```

```python
boston_data=pd.DataFrame(data)
```

```python
boston_data.head(10)
```

```python
boston_data.info()
```

```python
plt.hist(boston_data['MEDV'], bins=50, edgecolor='k')
plt.xlabel('Median Value')
plt.ylabel('Number of houses')
plt.title('Median value of owner occupied houses distribution')
```

```python
cols=['LSTAT','INDUS','NOX','RM','MEDV']
sns.pairplot(boston_data[cols])
```

```python
fig, x=plt.subplots()
sns.heatmap(np.corrcoef(boston_data[cols].values.T),annot=True)
x.set_xticklabels(cols)
x.set_yticklabels(cols)
plt.setp(x.get_xticklabels(),rotation=45,ha='right',rotation_mode='anchor')
plt.setp(x.get_yticklabels(),rotation=0)
```

```python
from sklearn.model_selection import train_test_split
X=boston_data[['RM']].values
y=boston_data['MEDV']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression().fit(np.array(X_train).reshape((-1,1)),y_train)
```

```python
model.score(np.array(X_test).reshape((-1,1)),y_test)
```

```python
X2=boston_data.iloc[:,:13]
y2=boston_data['MEDV']
X_train2,X_test2,y_train2,y_test2=train_test_split(X2,y2,test_size=0.33,random_state=42)
```

```python
model2=LinearRegression().fit(X_train2,y_train2)
```

```python
model2.score(X_test2,y_test2)
```

```python
y_train_pred=model2.predict(X_train2)
print(y_train2)
print(y_train_pred)
y_test_pred=model2.predict(X_test2)
print(y_test_pred)
```

```python
plt.scatter(y_train_pred,y_train_pred-y_train2,c="steelblue",marker="o",edgecolor='white',label="Training Data")
plt.scatter(y_test_pred,y_test_pred-y_test2,c="limegreen",marker="s",edgecolor="white",label="Test Data")
plt.xlabel("Pridicted Values")
plt.ylabel("Residuals")
plt.legend(loc='upper left')
plt.hlines(y=0,xmin=10,xmax=50,color='black',lw=2)
plt.xlim([-10,50])
plt.tight_layout
```

```python
from sklearn.metrics import mean_squared_error
print("MSE train: %.3f, test: %.3f" %(mean_squared_error(y_train2,y_train_pred),
                                      mean_squared_error(y_test2,y_test_pred)))
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=1.0)
ridge.fit(X_train2,y_train2)
ridge.score(X_test2,y_test2)
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=1.0)
lasso.fit(X_train2,y_train2)
lasso.score(X_test2,y_test2)
```

```python
from sklearn.linear_model import ElasticNet
en=ElasticNet(alpha=1.0,l1_ratio=0.5)
en.fit(X_train2,y_train2)
en.score(X_test2,y_test2)
```
